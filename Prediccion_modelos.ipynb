```python
import tensorflow as tf
import tensorflow_datasets as tfds

# Carga el conjunto de datos MNIST de TensorFlow Datasets
ds_test, info = tfds.load('mnist', split='test', shuffle_files=True, with_info=True)

# Normaliza las imágenes y convierte las etiquetas a enteros 
ds_test = ds_test.map(lambda x: (tf.cast(x['image'], tf.float32) / 255.0, x['label']))

# Define el modelo previamente entrenado
model = tf.keras.models.load_model('/content/gdrive/MyDrive/AtecneaYTB/modelo/modelo_con_validacion.h5')

# Evalúa el modelo en el conjunto de test
test_loss, test_acc = model.evaluate(ds_test.batch(32))

# Realiza predicciones en el conjunto de test y muestra algunas predicciones
for data, labels in ds_test.batch(5):
    predictions = model.predict(data)
    for i in range(5):
        print("Predicción: ", tf.argmax(predictions[i], axis=-1).numpy())
        print("Etiqueta real: ", labels[i].numpy())
```
