```python
import numpy as np
```

```python
# We will initialize randomly a matrix with the dimensions of the layers. 
# This will define the dimensions of our params, remember that the params are the variables that we dont set manually.
# They are intrinsic to the network
#matrix1 = np.random.randn(8,7) Random matrix (rows, columns)

def init_parameters(list_layers_dims): # we receive a list with the dim of the layers 
  parameters = {} # Creating an empy dict which will contain weights and bias and its values
  for layer in range(len(list_layers_dims)):
    parameters['W'+str(layer)] = np.random.randn(list_layers_dims[layer],list_layers_dims[layer-1]) # Setting the dims of weights matrix, this one has ('number of layers in the hidden layer', 'number of inputs')
    parameters['b'+str(layer)] = np.random.randn(list_layers_dims[layer],1) # The bias has the dims ('number of layers in the hidden layer',1)
  return parameters
```

```python
# Now we will define all the activation functions and their derivatives.

def relu(Z):
  return np.maximum(Z,0)

def softmax(Z):
  A = np.exp(Z) / sum(np.exp(Z))
  return A
```
